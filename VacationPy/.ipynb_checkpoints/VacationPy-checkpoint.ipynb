```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import gmaps
import requests
import os
from config import key_dict
from pprint import pprint

gkey = key_dict['maps']
```

```python
cities = "../output_data/cities.csv"
cities = pd.read_csv(cities)
cities
```

```python
gmaps.configure(api_key=gkey)
```

```python
locations = cities[['Lat', 'Lng']].astype(float)
humidity = cities['% Humidity'].astype(float)

figure_layout = {"width": "600px", "height": "400px"}

fig = gmaps.figure(layout=figure_layout)

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=humidity.max(), point_radius = 2)

fig.add_layer(heat_layer)

fig
```

```python
good_temp = cities['Max Temp'].between(292,303,inclusive=True)
good_humidity = cities['% Humidity'] < 85
clouds = cities['Cloudiness'].between(0,20,inclusive=True)
wind = cities['Wind Speed'] < 10
good_weather = cities[good_temp & good_humidity & clouds & wind] 
good_weather.head(10)
```

```python
df_hotel = good_weather.copy()

df_hotel["Hotel"] = ""
df_hotel
```

```python
url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
params = {
    'radius': 5000,
    'type': 'hotel',
    'key': gkey
}



for index, row in df_hotel.iterrows():
    
    #Get locations from dataframe
    Lat = row['Lat'] 
    Lng = row['Lng']
    location = f'{Lat},{Lng}'
    params['location']=location
    
    response = requests.get(url,params=params).json()
   
    results = response['results']
    
    try:
        
        print(f"The hotel name is {results[1]['name']}.")
        
        df_hotel.loc[index, 'Hotel Name'] = results[1]['name']
        
    except (KeyError, IndexError):
        print("No hotel found")
        
```
